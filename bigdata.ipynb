```python
#extraccion de datos con pandas 

#importacion de pandas
import pandas as pd
```

```python
#cargar el csv
df = pd.read_csv("centros-de-atencion-veterinaria.csv")
#mostrar las primeras cinco filas
df.head()
```

```python
#mostrar la cantidad de filas y columnas
print(df.shape)

#mostrar los nombres de las columnas
print(df.columns)

#mostrar los tipos de datos qe hay en cada columna
print(df.dtypes)

#vista
print(df.info())
```

```python
#cuantos datos son nulos
df.isnull()
```

```python
#mostrar cantidad de valores nulos por columna

print(df.isnull().sum())
```

```python
#estos datos ya estan limpios por los qe vamos directamente a estadisticas descriptivas

#mostrar estadisticas

print(df.describe())
```

```python
#visualizamos los datos con matplotlib
import matplotlib.pyplot as plt
```

```python
# Contamos cuántas veces aparece cada tipo de lugar
centros_por_lugar = df['lugar'].value_counts()

# Gráfico de barras
plt.figure(figsize=(10, 5))
plt.title("Cantidad de centros por lugar")
plt.bar(centros_por_lugar.index, centros_por_lugar.values, color='skyblue')
plt.xlabel("Lugar")
plt.ylabel("Cantidad de centros")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Agrupamos por día y contamos cuántos centros hay cada día
centros_por_dia = df['dias'].value_counts().sort_index()

# Tamaño de la figura
plt.figure(figsize=(10, 5))

# Título
plt.title("Cantidad de centros de atención por día")

# Pasamos los valores eje x (fechas) e y (cantidad)
plt.plot(centros_por_dia.index, centros_por_dia.values)

plt.xlabel("Fecha")
plt.ylabel("Cantidad de centros")
plt.xticks(rotation=20)
plt.grid()
plt.show()
```

```python
# Contamos cuántas veces aparece cada tipo de consulta
consultas_conteo = df['consultas'].value_counts()

# Gráfico de torta
plt.figure(figsize=(7, 7))
plt.pie(
    consultas_conteo.values,
    labels=consultas_conteo.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.Paired.colors  
)
plt.title('Distribución de Tipos de Consultas')
plt.axis('equal') 
plt.tight_layout()
plt.show()
```

```python
# Conteo de cada horario exacto
horarios_conteo = df['horarios'].value_counts()

# Gráfico de torta
plt.figure(figsize=(7, 7))
plt.pie(
    horarios_conteo.values,
    labels=horarios_conteo.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Distribución de Horarios Exactos')
plt.axis('equal')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```
